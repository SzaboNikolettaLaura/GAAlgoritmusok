````python
from itertools import product
DF2 = {
    "0000": 28, "0001": 26, "0010": 24, "0011": 18,
    "0100": 22, "0101":  6, "0110": 14, "0111":  0,
    "1000": 20, "1001": 12, "1010": 10, "1011":  2,
    "1100":  8, "1101":  4, "1110":  6, "1111": 30,
}

def bit_not(bits):
    return "".join("1" if b == "0" else "0" for b in bits)

f4 = {b: DF2[bit_not(b)] for b in DF2}

def create_bitstrings(bitek, include_star = False):
  if include_star:
    return ["".join(b) for b in product("01*", repeat=bitek)]
  return ["".join(b) for b in product("01", repeat=bitek)]

def flip_zeros_to_ones(schema):
  return ''.join(['1' if letter == '0' else letter for letter in schema])

def avg_fitness_of_schema(schema, bitek):
  good = []
  for string in create_bitstrings(bitek):
    matches = True
    for i in range(bitek):
      if schema[i] != '*' and schema[i] != string[i]:
        matches = False
        break
    if matches:
      good.append(string)
  if len(good) == 0:
    return 0
  return sum([f4[string] for string in good]) / len(good)

def main():

  osszes = create_bitstrings(4)

  for block in osszes:
    print(block, f4[block])

  maxfit = max(f4[x] for x in osszes)
  bestp = [x for x in osszes if f4[x] == maxfit]
  print("Globális maximum érték:", maxfit)
  print("Globális maximum helye(i):", bestp)

  violations = []

  for schema in create_bitstrings(4, True):
    if schema == "****":
      continue
    if '0' not in schema:
      continue
    if '*' not in schema:
      continue

    flip = flip_zeros_to_ones(schema)
    avga = avg_fitness_of_schema(schema, 4)
    avgb = avg_fitness_of_schema(flip, 4)

    if not avgb > avga:
      violations.append((schema, flip, avga, avgb))

  if len(violations) == 0:
    print("Minden vizsgált sémára teljesül:")
    print("  a 0→1 csere növeli a szkéma átlagos fitnesszét")
  else:
    print("Találtunk ellenpéldá(kat):")
    for schema, flip, avga, avgb in violations:
      print(schema, "->", flip, " - ", avga, "->", avgb)

main()
````

````python
from itertools import product

DF2 = {
    "0000": 28, "0001": 26, "0010": 24, "0011": 18,
    "0100": 22, "0101":  6, "0110": 14, "0111":  0,
    "1000": 20, "1001": 12, "1010": 10, "1011":  2,
    "1100":  8, "1101":  4, "1110":  6, "1111": 30,
}

def f8(x):
  y = bit_not(x)
  b1 = y[0:4]
  b2 = y[4:8]
  return DF2[b1] + DF2[b2]


def bit_not(bits):
    return "".join("1" if b == "0" else "0" for b in bits)


def create_bitstrings(bitek, include_star = False):
  if include_star:
    return ["".join(b) for b in product("01*", repeat=bitek)]
  return ["".join(b) for b in product("01", repeat=bitek)]

def flip_zeros_to_ones(schema):
  return ''.join(['1' if letter == '0' else letter for letter in schema])

def avg_fitness_of_schema(schema, bitek):
  good = []
  for string in create_bitstrings(bitek):
    matches = True
    for i in range(bitek):
      if schema[i] != '*' and schema[i] != string[i]:
        matches = False
        break
    if matches:
      good.append(string)
  if len(good) == 0:
    return 0
  return sum([f8(string) for string in good]) / len(good)

def main():

  osszes = create_bitstrings(8)

  for block in osszes:
    print(block, f8(block))

  maxfit = max(f8(x) for x in osszes)
  bestp = [x for x in osszes if f8(x) == maxfit]
  print("Globális maximum érték:", maxfit)
  print("Globális maximum helye(i):", bestp)

  violations = []

  for schema in create_bitstrings(8, True):
    if schema == "********":
      continue
    if '0' not in schema:
      continue
    if '*' not in schema:
      continue

    flip = flip_zeros_to_ones(schema)
    avga = avg_fitness_of_schema(schema, 8)
    avgb = avg_fitness_of_schema(flip, 8)

    if not avgb > avga:
      violations.append((schema, flip, avga, avgb))

  if len(violations) == 0:
    print("Minden vizsgált sémára teljesül:")
    print("  a 0→1 csere növeli a szkéma átlagos fitnesszét")
  else:
    print("Találtunk ellenpéldá(kat):")
    for schema, flip, avga, avgb in violations:
      print(schema, "->", flip, " - ", avga, "->", avgb)

main()
````
